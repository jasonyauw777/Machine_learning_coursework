```python
# import pandas library
# read csv file

import pandas as pd

path = "/content/drive/MyDrive/Colab Notebooks/iris.csv"

iris_df = pd.read_csv(path)
```

```python
iris_df
```

```python
iris_df.loc[len(iris_df.index)] = [5.1,	3.5,	1.4,	0.2,	'Iris-setosa']

display(iris_df)
```

```python
iris_df = iris_df.rename(columns={"5.1": "sepal length", "3.5": "sepal width", 
                                  "1.4": "petal length", "0.2": "petal width", 
                                  "Iris-setosa": "class"})
```

```python
display(iris_df)
```

```python
iris_df = iris_df.dropna(axis=1)
```

```python
for i in iris_df:
  print(i, iris_df.isna().values.any())
```

```python
display(iris_df)
```

## **KNN CLASSIFICATION**

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(iris_df, test_size = 0.3)
```

```python
KNN = KNeighborsClassifier(n_neighbors=5).fit(train[["sepal length", "sepal width",
                                                     "petal length","petal width"]]
                                                    , train["class"])
```

```python
flower_classification_test = KNN.predict(test[["sepal length", "sepal width",
                                               "petal length","petal width"]])

flower_classification_train = KNN.predict(train[["sepal length", "sepal width",
                                                 "petal length","petal width"]])

accuracy_testdata = accuracy_score(test["class"], flower_classification_test)

accuracy_traindata = accuracy_score(train["class"], flower_classification_train)

print(" Accuracy Test Data : %.2f" % accuracy_testdata)

print(" Accuracy Train Data : %.2f" % accuracy_traindata)
```

## **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, f1_score
```

```python
LGR = LogisticRegression(C = 0.03, solver = 'liblinear').fit(train[["sepal length", "sepal width",
                                                                    "petal length","petal width"]], train["class"])
```

```python
flower_classification_test2 = LGR.predict(test[["sepal length", "sepal width",
                                                "petal length","petal width"]])

flower_classification_train2 = LGR.predict(train[["sepal length", "sepal width",
                                                  "petal length","petal width"]])

accuracy_testdata2 = accuracy_score(test["class"], flower_classification_test2)

accuracy_traindata2 = accuracy_score(train["class"], flower_classification_train2)

print(" Accuracy Test Data : %.2f" % accuracy_testdata2)

print(" Accuracy Train Data : %.2f" % accuracy_traindata2)
```

## **SVM**

```python
from sklearn import svm
```

```python
SVM = svm.SVC().fit(train[["sepal length", "sepal width",
                           "petal length","petal width"]], train["class"])
```

```python
flower_classification_test3 = SVM.predict(test[["sepal length", "sepal width",
                                                "petal length","petal width"]])

flower_classification_train3 = SVM.predict(train[["sepal length", "sepal width",
                                                  "petal length","petal width"]])

accuracy_testdata3 = accuracy_score(test["class"], flower_classification_test3)

accuracy_traindata3 = accuracy_score(train["class"], flower_classification_train3)

print(" Accuracy Test Data : %.2f" % accuracy_testdata3)

print(" Accuracy Train Data : %.2f" % accuracy_traindata3)
```

Overall, all the machine learning models are not underfitting or overfitting.

The KNN model performs the best in terms of accuracy on both test and training data with an accuracy score of 0.98 and 0.96 respectively. 

The logistic regression model performs the worst out of all 3 models with an accuracy score of 0.93 and 0.90 on test and training data respectively.

The SVM model performs just as well as KNN on the test data, but performs worse on the training data with an accuracy score of 0.93 compared to the KNN's score of 0.96.

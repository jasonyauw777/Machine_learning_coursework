**1**. Dicrete and Continous random variables are two types of quantitive variables. Dicrete random variables represents counts and can only take certain values (eg, 1 human, not 1/2 human). Continous random variables represents measurable amounts and takes values from a certain interval. 

In the data set, the Discrete random variables are **id**, **host_id**,  **minimum_nights**, **number_of_reviews**, **calculated_host_listings_count** and **availability_365**. 

The continous random variables are **latitude**, **longtitude** and **reviews_per_month**, **price**, **last_review**.


```
import pandas as pd

path = '/content/drive/MyDrive/Colab Notebooks/listings.csv'
dl = pd.read_csv(path)

dl.head()
```

2. Pre-process the data, you can replace NaN values with zero and the mean of the variable.

```
print(dl.head())

print(dl.isna().values.any())
newDataListingsDrop = dl.dropna()
newDataListingsFill = dl.fillna(0)
print("Original Length:", len(dl))
print("Dropped Length:", len(newDataListingsDrop))
print("Replaced Length:", len(newDataListingsFill))
```

```
for i in dl:
  print(i, dl[i].isna().values.any())

print("\n")

newDataListingsFill2 = dl
for i in dl:
  if dl[i].dtypes == object :
    newDataListingsFill2[i] = dl[i].fillna("Not Available")
  else :
    mean = dl[i].mean()
    newDataListingsFill2[i] = dl[i].fillna(mean)

print(newDataListingsFill2)
```

3. Mean and Variance for each continous and discrete random variables. For data listings where NaN values are replaced with 0 and mean.


```
print("NaN values replaced by 0\n")
for i in newDataListingsFill:
  if newDataListingsFill[i].dtypes != object :
     print("Mean value of ", i, "in newDataListingsFill is ", newDataListingsFill[i].mean())
     print("Variance of ", i, "in newDataListingsFill is ", newDataListingsFill[i].var())
print("\n")

print("NaN values replaced by mean\n")
for i in newDataListingsFill2:
  if newDataListingsFill[i].dtypes != object :
    print("Mean value of ", i, "in newDataListingsFill2 is ", newDataListingsFill2[i].mean())
    print("Variance of ", i, "in newDataListingsFill2 is ", newDataListingsFill2[i].var())
print("\n")
```

4. In this particular case for both Data Sets where NaN values are replaced by 0 and mean values respectively. There is no significant difference in terms of variance in all variables between both data sets.

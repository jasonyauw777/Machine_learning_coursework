```python
import pandas as pd

path = '/content/drive/MyDrive/Colab Notebooks/listings.csv'
dl = pd.read_csv(path)

dl.head()
```

**Classification - KNN**

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

for i in dl:
  print(i, dl[i].dtypes)
  if dl[i].isna().values.any() and dl[i].dtypes != object:
      dl[i] = dl[i].fillna(dl[i].mean())

train, test = train_test_split(dl, test_size=0.2)
```

```python
KNN = KNeighborsClassifier(n_neighbors = 3).fit(train[["latitude", "longitude"]], train["price"])
```

```python
newClassification = KNN.predict([[1.44255, 103.79580],[1.44246, 103.79667]])
print(newClassification)

classification = KNN.predict(test[["latitude", "longitude"]])
accuracy = accuracy_score(test["price"], classification)
MAE = mean_absolute_error(test["price"], classification)
MSE = mean_squared_error(test["price"], classification)

print(" ACC : %.2f" % accuracy)
print(" MAE : %.2f" % MAE)
print(" MSE : %.2f" % MSE)
```

```python
Ks = 10
accuracy = np.zeros((Ks-1))
ConfustionMx = [];
for n in range(1, Ks):    
    KNN = KNeighborsClassifier(n_neighbors = n).fit(train[["latitude", "longitude"]], train["price"])  
    classification = KNN.predict(test[["latitude", "longitude"]])
    accuracy[n - 1] = accuracy_score(test["price"], classification)
    
print("Best  ACC : %.2f" % accuracy.max(), ", with k = ", accuracy.argmax() + 1)
```

**Classification - SVM**

```python
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, f1_score
```

```python
SVM = svm.SVC().fit(train[["latitude", "longitude"]], train["price"])
```

```python
newClassification = SVM.predict([[1.44255, 103.79580],[1.44246, 103.79667]])
print(newClassification)

classification = SVM.predict(test[["latitude", "longitude"]])
accuracy = accuracy_score(test["price"], classification)
MAE = mean_absolute_error(test["price"], classification)
MSE = mean_squared_error(test["price"], classification)
F1  = f1_score(test["price"], classification, average='weighted') 
cnf_matrix = confusion_matrix(test["price"], classification, labels=[83,69])


print(" ACC : %.2f" % accuracy)
print(" MAE : %.2f" % MAE)
print(" MSE : %.2f" % MSE)
print(" F1  : %.2f" % F1)
print (classification_report(test["price"], classification))
print(cnf_matrix)
```
